## Logistic regression decision boundary with 3 classes

<iframe width="560" height="315" src="https://www.youtube.com/embed/BDfoP8CAISE" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>

```python
import altair as alt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins").dropna()
```

```python
df.columns
```

```python
cols = ['bill_length_mm', 'bill_depth_mm']
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color="species"
)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(df[cols], df["species"])
```

```python
df["pred"] = clf.predict(df[cols])
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color="pred"
)
```

```python
df["species"] == df["pred"]
```

```python
df["correct"] = (df["species"] == df["pred"])
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color="pred"
)
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color="correct"
)
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color=alt.condition(alt.datum.correct, alt.value("black"), alt.value("red"))
)
```

```python
df.correct.unique()
```

```python
alt.Chart(df).mark_circle().encode(
    x=alt.X(cols[0], scale=alt.Scale(zero=False)),
    y=alt.Y(cols[1], scale=alt.Scale(zero=False)),
    color=alt.Color("correct", scale=alt.Scale(domain=[True, False], range=["black", "red"]))
)
```
